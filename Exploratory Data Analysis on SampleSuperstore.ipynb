## **Univariate Analysis**

### **Analyzing the segment of people in the store**

```python
no_segment = dataset['Segment'].value_counts()
print(no_segment)
```

```python
percent_segment = no_segment / len(dataset['Segment']) * 100
print(percent_segment)
```

```python
percent_segment.plot(kind = 'bar', color = 'red')
print("Percentage Of Each Segment")
```

### Conclusion:

1.   More than 50% of the people in the Store are Consumers
2.   Around 30% of the people in the Store are in the Corporate Segment
3.   Around 20% of the people in the Store belong to Home Office 



### **Analyzing the region of orders**

```python
no_region = dataset['Region'].value_counts()
print(no_region)
```

```python
percent_region = no_region / len(dataset['Region']) * 100
print(percent_region)
```

```python
percent_region.plot(kind ='bar')
```

## Conclusion:

The least orders are from South whereas the maximum orders are from West

### **Analyzing the Ship Mode in the store**

```python
no_ShipMode = dataset['Ship Mode'].value_counts()
print(no_ShipMode)
```

```python
percent_ShipMode = no_ShipMode / len(dataset['Ship Mode']) * 100
print(percent_ShipMode)
```

```python
percent_ShipMode.plot(kind = 'bar', color = 'cyan')
```

### Conclusion:


1.   The Standard Class has the most no, of modes i.e 60%
2.   Less No. of deliveries reach the same day i.e 5%





### **Analyzing Categories of orders**

```python
no_category = dataset['Category'].value_counts()
print(no_category)
```

```python
percent_category = no_category / len(dataset['Category']) * 100
myexplode = [0.2, 0, 0]
print(percent_category)
```

```python
percent_category.plot(kind = 'pie', explode = myexplode, shadow = "true")
plt.legend(title = 'Categories Of Orders:')
```

### Conclusion:


1.   Office Supplies contains 60% orders

2.   20% is of Furniture and 20% is of Technology

### **Plotting Subcategories**

### Conclusion:

1.   Blinder has been bought maximum followed by the paper
2.   Copiers have been bought least frequently



### **Counting number of orders from each state**

```python
print(dataset['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=dataset['State'])
plt.xticks(rotation=90)
plt.show()
```

### **Analyzing Sales and Profit per Segment**

```python
fig, ax = plt.subplots()
colors={'Consumer':'yellow','Corporate':'red','Home Office':'green'}
ax.scatter(dataset['Sales'],dataset['Profit'],c=dataset['Segment'].apply(lambda x: colors[x]))
plt.show()
```

```python
x=data.groupby(dataset['Segment']).agg('sum')
print(x)

x['Sales'].plot(kind='bar', color = 'red')
```

```python
x['Profit'].plot(kind='bar')
```

```python
dataset.pivot_table(values='Sales',index='Segment',columns='Discount',aggfunc='median')
```

```python
dataset.pivot_table(values='Profit',index='Segment',columns='Discount',aggfunc='median')
```

The consumer segment has the maximum Sales and Profit
